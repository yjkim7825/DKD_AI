# 장비별 probe 이름 비교
같은 bulk라도 장비(플랫폼)마다 이름 형식이 다름 → 어레이는 probe→gene 매핑 필요.

```python
import pandas as pd, gzip
B = "../01_preprocessing/bulk/output/02_normalized"
def probes(g, n=6):
    ids=[]
    with gzip.open(f"{B}/{g}_series_matrix.txt.gz","rt") as f:
        on=False
        for l in f:
            if l.startswith('"ID_REF"'): on=True; continue
            if on:
                if l.startswith("!"): break
                ids.append(l.split("\t")[0].strip('"'))
                if len(ids)>=n: break
    return ids
rna=[]
with gzip.open("../../data/1-4. GSE142025_RAW/GSM4217781_A11A.txt.gz","rt") as f:
    next(f)
    for l in f:
        rna.append(l.split("\t")[0])
        if len(rna)>=6: break
cmp = pd.DataFrame({
  "GSE30529 (HG-U133A2)": probes("GSE30529"),
  "GSE96804 (HTA-2.0)":   probes("GSE96804"),
  "GSE142025 (RNA-seq)":  rna})
print("장비별 probe/유전자 이름 (각 앞 6개) — 어레이는 _at/_st probe, RNA-seq는 유전자 이름")
cmp
```

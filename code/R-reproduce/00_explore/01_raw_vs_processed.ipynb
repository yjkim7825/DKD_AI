# 전처리: 원본 → 결과 (5개 각각) + 정규화 과정

데이터마다 **원본 → 전처리 결과**. 큰 원본은 상위 몇 행만 미리보기(전체 차원은 shape로 표시).
정규화가 실제 실행되는 **GSE142025는 정규화 전/후**도 표시.

```python
import pandas as pd, os, gzip, glob
from io import StringIO
pd.set_option('display.max_columns',18)
OLD='../../data/processed'; RAW='../../data'; BASE='../01_preprocessing/bulk/output'
# 어레이 원본: 전체 행수는 세고, 미리보기는 상위 n행만 (빠르게)
def series_preview(g,n=8):
    with gzip.open(f'{OLD}/{g}_series_matrix.txt.gz','rt') as f:
        L=[next(f) for _ in range(200)]  # 헤더+앞부분만
    s=next(i for i,l in enumerate(L) if l.startswith('"ID_REF"'))
    prev=pd.read_csv(StringIO(''.join(L[s:s+1+n])),sep='\t',index_col=0)
    total=sum(1 for _ in gzip.open(f'{OLD}/{g}_series_matrix.txt.gz','rt'))-s-1
    return prev,total
```

## 정규화 과정 (GSE142025)
RNA-seq는 log2+정규화 실제 수행. **샘플별 평균**이 정규화 후 균일해짐.
(어레이는 RMA가 이미 정규화 → 스킵)

```python
before=pd.read_csv(f'{BASE}/01_genematrix/GSE142025.genematrix.txt',sep='\t',index_col=0)
after =pd.read_csv(f'{BASE}/02_normalized/GSE142025.normalized.txt',sep='\t',index_col=0)
cmp=pd.DataFrame({'정규화 전 평균':before.mean().round(3).values[:8],
                  '정규화 후 평균':after.mean().round(3).values[:8]}, index=list(before.columns[:8]))
print('앞 8개 샘플 평균: 전은 들쭉날쭉 → 후는 균일'); cmp
```

## 자동 대조 — 새 vs 기존

```python
def gm(c): return {x.split('_')[0]:x for x in c}
rows=[]
for g in ['GSE30529','GSE96804','GSE104948','GSE104954','GSE142025']:
    n=pd.read_csv(f'{BASE}/01_genematrix/{g}.genematrix.txt',sep='\t',index_col=0)
    o=pd.read_csv(f'{OLD}/{g}.labeled.txt',sep='\t',index_col=0)
    nm,om=gm(n.columns),gm(o.columns); cm=sorted(set(nm)&set(om))
    ck=lambda x:('-' if (x not in n.index or x not in o.index or not cm) else ('O' if abs(n.loc[x,nm[cm[0]]]-o.loc[x,om[cm[0]]])<0.01 else 'X'))
    rows.append([g,f'{n.shape[0]}×{n.shape[1]}',f'{o.shape[0]}×{o.shape[1]}',ck('FN1'),ck('ALDH2')])
pd.DataFrame(rows,columns=['데이터','새','기존','FN1','ALDH2'])
```

## 논문(Table S1) 대조 — 샘플 수

※ 논문 S1은 **샘플 수만** 제공 (유전자 수는 없음).
→ 샘플 수로 대조. (유전자 수는 논문에 없어 대조 불가)


```python
import pandas as pd, os
from collections import Counter
B='../01_preprocessing/bulk/output'; 
paper={'GSE96804':'DKD 41 / Ctrl 20','GSE104948':'DKD 12 / Ctrl 21','GSE104954':'DKD 17 / Ctrl 21',
       'GSE142025':'Early 6 / Late 21 / Ctrl 9','GSE30529':'DKD 10 / Ctrl 12'}
rows=[]
for g in ['GSE96804','GSE104948','GSE104954','GSE142025','GSE30529']:
    df=pd.read_csv(f'{B}/02_normalized/{g}.normalized.txt',sep='\t',index_col=0)
    c=Counter(x.split('_')[-1] for x in df.columns)
    ours=' / '.join(f'{k} {v}' for k,v in c.items())
    match='일치' if sum(c.values())==sum(int(x.split()[1]) for x in paper[g].split(' / ')) else '확인'
    rows.append([g, paper[g], ours, match, f'{df.shape[0]} (논문 N/A)'])
pd.DataFrame(rows,columns=['데이터','논문 샘플(S1)','우리 샘플','샘플수 일치','전처리후 유전자'])
```
